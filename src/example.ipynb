```python
# Loading first 5 frames of a clip
m = []
cap = cv.VideoCapture('original/sitting-man.mp4')
for i in range(5):
    ret, frame = cap.read()
    m.append(cv.cvtColor(frame, cv.COLOR_BGR2GRAY))
m =  np.array(m)
```

```python
plt.imshow(m[0], cmap='gray', vmin=0, vmax=255)
```

```python
# Detecting rain/snow from movie frames
# The angle (100) is estimated manually
rain = detect(m, np.pi*100/180)
```

```python
# Plotting original scene and detected rain/snow
frame_number = 0
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.title("Original")
plt.imshow(m[frame_number], cmap='gray', vmin=0, vmax=255)
plt.subplot(122)
plt.title("Rain/Snow")
plt.imshow(rain[frame_number], cmap='gray')
```

```python
# Removing rain/snow from movie frames
clean = remove(m, rain, 3)
```

```python
frame_number = 0
plt.figure(figsize=(20, 10))
plt.subplot(131)
plt.title("Original")
plt.imshow(m[frame_number], cmap='gray', vmin=0, vmax=255)
plt.subplot(132)
plt.title("Clean")
plt.imshow(clean[frame_number], cmap='gray', vmin=0, vmax=255)
plt.subplot(133)
plt.title("Rain/Snow")
plt.imshow(rain[frame_number], cmap='gray')
```
